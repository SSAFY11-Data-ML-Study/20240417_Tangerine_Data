```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

# timestamp 컬럼의 월별로 그룹화합니다.
df['month'] = df['timestamp'].dt.to_period('M')  # 월별로 기간(Period)로 변환

# 월별로 groupby합니다.
grouped = df.groupby(['item', 'month']).agg({
    'price(원/kg)': 'mean'
}).reset_index()

# 결과를 출력합니다.
print(grouped)
```

```python
grouped_tg = grouped[grouped['item'] == 'TG']
grouped_bc = grouped[grouped['item'] == 'BC']
grouped_cb = grouped[grouped['item'] == 'CB']
grouped_cr = grouped[grouped['item'] == 'CR']
grouped_rd = grouped[grouped['item'] == 'RD']
```

```python
def plot_price_by_item(grouped_item, item_name):
    plt.figure(figsize=(10, 6))
    plt.plot(grouped_item['month'].dt.to_timestamp(), grouped_item['price(원/kg)'], marker='o', label=item_name)
    plt.xlabel('Month')
    plt.ylabel('Price')
    plt.title(f'Monthly Price Change for item {item_name}')
    plt.grid(True)
    plt.legend()
    plt.show()

grouped_tg = grouped[grouped['item'] == 'TG']
grouped_bc = grouped[grouped['item'] == 'BC']
grouped_cb = grouped[grouped['item'] == 'CB']
grouped_cr = grouped[grouped['item'] == 'CR']
grouped_rd = grouped[grouped['item'] == 'RD']

plot_price_by_item(grouped_tg, 'TG')
```

### 상품별로 나누어서 해보자


```python
df = pd.read_csv('./jeju/train.csv')
df_test = pd.read_csv('./jeju/test.csv')
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 'year', 'month', 'day' 열 추가
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.dayofweek
```

```python
df = df.drop(['ID', 'timestamp', 'supply(kg)'], axis=1)
```

```python
df['location'] = df['location'].apply(lambda x: 1 if x == 'J' else 0)
df_test['location'] = df_test['location'].apply(lambda x: 1 if x == 'J' else 0)
```

```python
label_encoder = LabelEncoder()
df_test['corporation'] = label_encoder.fit_transform(df_test['corporation'])
df['corporation'] = label_encoder.fit_transform(df['corporation'])
```

```python
df_tg = df[df['item'] =='TG']
df_bc = df[df['item'] =='BC']
df_rd = df[df['item'] =='RD']
df_cb = df[df['item'] =='CB']
df_cr = df[df['item'] =='CR']
```

```python
df_test['timestamp'] = pd.to_datetime(df_test['timestamp'])

# 'year', 'month', 'day' 열 추가
df_test['year'] = df_test['timestamp'].dt.year
df_test['month'] = df_test['timestamp'].dt.month
df_test['day'] = df_test['timestamp'].dt.dayofweek
```

```python
df_test = df_test.drop(['timestamp'], axis=1)
```

```python
test_tg = df_test[df_test['item'] =='TG']
test_bc = df_test[df_test['item'] =='BC']
test_rd = df_test[df_test['item'] =='RD']
test_cb = df_test[df_test['item'] =='CB']
test_cr = df_test[df_test['item'] =='CR']
```

```python
test_tg_id = test_tg['ID']
test_tg = test_tg.drop('ID',axis=1)
test_bc_id = test_bc['ID']
test_bc = test_bc.drop('ID',axis=1)
test_rd_id = test_rd['ID']
test_rd = test_rd.drop('ID',axis=1)
test_cb_id = test_cb['ID']
test_cb = test_cb.drop('ID',axis=1)
test_cr_id = test_cr['ID']
test_cr = test_cr.drop('ID',axis=1)
```

```python
df_tg = df_tg.drop('item',axis=1)
test_tg = test_tg.drop('item',axis=1)
df_rd = df_rd.drop('item',axis=1)
test_rd = test_rd.drop('item',axis=1)
df_cb = df_cb.drop('item',axis=1)
test_cb = test_cb.drop('item',axis=1)
df_cr = df_cr.drop('item',axis=1)
test_cr = test_cr.drop('item',axis=1)
df_bc = df_bc.drop('item',axis=1)
test_bc = test_bc.drop('item',axis=1)
```

```python
x_tg = df_tg.drop('price(원/kg)', axis=1)
y_tg = df_tg['price(원/kg)']
x_rd = df_rd.drop('price(원/kg)', axis=1)
y_rd = df_rd['price(원/kg)']
x_cb = df_cb.drop('price(원/kg)', axis=1)
y_cb = df_cb['price(원/kg)']
x_cr = df_cr.drop('price(원/kg)', axis=1)
y_cr = df_cr['price(원/kg)']
x_bc = df_bc.drop('price(원/kg)', axis=1)
y_bc = df_bc['price(원/kg)']
```

```python
x_tg_train, x_tg_test, y_tg_train, y_tg_test = train_test_split(x_tg, y_tg, test_size=0.4, random_state=42)
x_rd_train, x_rd_test, y_rd_train, y_rd_test = train_test_split(x_rd, y_rd, test_size=0.4, random_state=42)
x_cb_train, x_cb_test, y_cb_train, y_cb_test = train_test_split(x_cb, y_cb, test_size=0.4, random_state=42)
x_cr_train, x_cr_test, y_cr_train, y_cr_test = train_test_split(x_cr, y_cr, test_size=0.4, random_state=42)
x_bc_train, x_bc_test, y_bc_train, y_bc_test = train_test_split(x_bc, y_bc, test_size=0.4, random_state=42)
```

```python
model_tg = CatBoostRegressor(cat_features=['location', 'corporation', 'month', 'year', 'day'], verbose=50, iterations=2000)
model_rd = CatBoostRegressor(cat_features=['location', 'corporation', 'month', 'year', 'day'], verbose=50, iterations=2000)
model_cb = CatBoostRegressor(cat_features=['location', 'corporation', 'month', 'year', 'day'], verbose=50, iterations=2000)
model_cr = CatBoostRegressor(cat_features=['location', 'corporation', 'month', 'year', 'day'], verbose=50, iterations=2000)
model_bc = CatBoostRegressor(cat_features=['location', 'corporation', 'month', 'year', 'day'], verbose=50, iterations=2000)
```

```python
model_tg.fit(x_tg_train, y_tg_train)
```

```python
model_rd.fit(x_rd_train, y_rd_train)
model_bc.fit(x_bc_train, y_bc_train)
model_cb.fit(x_cb_train, y_cb_train)
model_cr.fit(x_cr_train, y_cr_train)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# 모델 예측값 생성
y_tg_pred = model_tg.predict(x_tg_test)
y_rd_pred = model_rd.predict(x_rd_test)
y_cb_pred = model_cb.predict(x_cb_test)
y_cr_pred = model_cr.predict(x_cr_test)
y_bc_pred = model_bc.predict(x_bc_test)
# RMSE 계산
rmse_tg = np.sqrt(mean_squared_error(y_tg_test, y_tg_pred))
rmse_rd = np.sqrt(mean_squared_error(y_rd_test, y_rd_pred))
rmse_cb = np.sqrt(mean_squared_error(y_cb_test, y_cb_pred))
rmse_cr = np.sqrt(mean_squared_error(y_cr_test, y_cr_pred))
rmse_bc = np.sqrt(mean_squared_error(y_bc_test, y_bc_pred))
print("RMSE for model_tg:", rmse_tg)
print("RMSE for model_rd:", rmse_rd)
print("RMSE for model_cb:", rmse_cb)
print("RMSE for model_cr:", rmse_cr)
print("RMSE for model_bc:", rmse_bc)
```

```python
tg_pred = model_tg.predict(test_tg)
rd_pred = model_rd.predict(test_rd)
cb_pred = model_cb.predict(test_cb)
cr_pred = model_cr.predict(test_cr)
bc_pred = model_bc.predict(test_bc)
```
